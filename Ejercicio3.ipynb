```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/MiguelUTEC26/correlacion-estructurado/refs/heads/main/data/raw/Ejercicio%203.csv")
```

```python
print(df.head())
print(df.shape)
print(df.info())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
variables_numericas = ["Weight", "Cholesterol"]

for columna in variables_numericas:
    media = df[columna].mean()
    desviacion = df[columna].std()

    limite_inferior = media - 2 * desviacion
    limite_superior = media + 2 * desviacion

    atipicos = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]
    print(columna, limite_inferior, limite_superior, len(atipicos))
```

```python
from sklearn.cluster import KMeans

X = df[["Weight", "Cholesterol", "Gender"]]

modelo_k2 = KMeans(n_clusters=2, random_state=0, n_init=10)
df["cluster_k2"] = modelo_k2.fit_predict(X)

centroides_k2 = pd.DataFrame(modelo_k2.cluster_centers_, columns=X.columns)
print(centroides_k2)
```

```python
modelo_k4 = KMeans(n_clusters=4, random_state=0, n_init=10)
df["cluster_k4"] = modelo_k4.fit_predict(X)

centroides_k4 = pd.DataFrame(modelo_k4.cluster_centers_, columns=X.columns)
print(centroides_k4)
print(df["cluster_k4"].value_counts())
```

```python
grupo_riesgo = df[df["cluster_k4"] == 1]

print(grupo_riesgo[["Weight", "Cholesterol", "Gender"]].describe())
print(grupo_riesgo[["Weight", "Cholesterol"]].min())
```

```python
# Filtrar empleados/personas asignadas al clúster de mayor riesgo
personas_riesgo = df[df["cluster_k4"] == 1]

# Guardar el resultado para revisión o uso posterior
personas_riesgo.to_csv("personas_riesgo_cluster_kmeans.csv", index=False)

# Criterio mínimo observado en el clúster de riesgo
peso_minimo = personas_riesgo["Weight"].min()
colesterol_minimo = personas_riesgo["Cholesterol"].min()

print(peso_minimo, colesterol_minimo)
```
